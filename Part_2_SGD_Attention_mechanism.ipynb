```python
# Input features (X) and target values (y) for two examples
X = np.array([1, 2])  # Example feature values
y = np.array([3, 4])  # Example target values
```

```python
# Calculate predictions
predictions = X * weight
```

```python
# Compute loss for each example
losses = (predictions - y) ** 2
```

```python
# Update weights using SGD for the first example
gradient = 2 * (predictions[0] - y[0]) * X[0]
weight -= learning_rate * gradient
```

```python
# Update weights using SGD for the second example
gradient = 2 * (predictions[1] - y[1]) * X[1]
weight -= learning_rate * gradient
```

Attention Mechanism

```python
import torch
import torch.nn.functional as F
from transformers import BertTokenizer, BertModel
```

```python
# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Input sentences
sentence1 = "The bank is closed today."
sentence2 = "She sat by the river bank and enjoyed the view."
```

```python
# Tokenize and encode the sentences
inputs1 = tokenizer(sentence1, return_tensors='pt', padding=True, truncation=True)
inputs2 = tokenizer(sentence2, return_tensors='pt', padding=True, truncation=True)
```

```python
# Get the word embeddings from BERT model
with torch.no_grad():
    outputs1 = model(**inputs1)
    outputs2 = model(**inputs2)
```

```python
# Get the word embeddings for each token in the sentences
embeddings1 = outputs1.last_hidden_state
embeddings2 = outputs2.last_hidden_state
```

```python
# Compute attention weights using dot product attention
attention_weights = torch.matmul(embeddings1, embeddings2.transpose(1, 2))
attention_weights = F.softmax(attention_weights, dim=-1)
```

```python
# Weighted sum of word embeddings in the second sentence
weighted_sum = torch.matmul(attention_weights, embeddings2)
```

```python
# Convert attention weights and weighted sum to numpy arrays
attention_weights = attention_weights.cpu().numpy()
weighted_sum = weighted_sum.cpu().numpy()
```

```python
# Display attention weights and weighted sum
print("Attention Weights:")
print(attention_weights)
print("\nWeighted Sum:")
print(weighted_sum)
```
